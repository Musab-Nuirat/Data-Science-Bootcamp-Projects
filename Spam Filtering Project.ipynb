```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

# Load the data
data = pd.read_csv('emails.csv') 

# Split into features (X) and target labels (y)
X = data['text']
y = data['spam']

# Perform text vectorization
vectorizer = TfidfVectorizer(max_features=1000)  
X_vectorized = vectorizer.fit_transform(X)

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_vectorized, y, test_size=0.2, random_state=42)

# Initialize the XGBoost classifier with suggested hyperparameters
clf = XGBClassifier(
    learning_rate=0.05, 
    n_estimators=500,   
    max_depth=3,        
    subsample=0.9,      
)
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

### Second Solution

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import ShuffleSplit
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

# Load the data
data = pd.read_csv('emails.csv')  


# Split into features (X) and target labels (y)
X = data['text']
y = data['spam']

# Perform text vectorization
vectorizer = TfidfVectorizer(max_features=1000)  # Adjust the number of features as needed
X_vectorized = vectorizer.fit_transform(X)

# Convert the vectorized features into a DataFrame
X_df = pd.DataFrame(X_vectorized.toarray(), columns=vectorizer.get_feature_names_out())

# Merge X_df with the target labels (y) into a single DataFrame
data = pd.concat([X_df, y], axis=1)

# data = X_vectorized
# Splitting the data
n = len(data)
train = data[0:(n//10)*8]
test = data[(n//10)*8:]
y_train0 = train['spam']
y_test0 = test['spam']
X_train0 = train.drop('spam', axis=1)
X_test0 = test.drop('spam', axis=1)
X = np.array(X_train0)
y = np.array(y_train0)

# Initialize the XGBoost classifier
clf = XGBClassifier(max_depth=3, n_estimators=1000, learning_rate=0.01)

# Initialize ShuffleSplit
ss = ShuffleSplit(n_splits=5, train_size=0.8, test_size=0.2, random_state=0)

# Training and evaluating the model
for train_index, test_index in ss.split(X):
    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = y[train_index], y[test_index]

    clf.fit(X_train, Y_train)
    print(clf.score(X_test, Y_test))

# Making predictions on the test set
pred0 = clf.predict(np.array(X_test0))
PRED = pred0
ANS = y_test0
accuracy = accuracy_score(ANS, PRED)
print("Final Accuracy:", accuracy)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

# Load the data
data = pd.read_csv('email_v2.csv') 

data.head(5)

# Split into features (X) and target labels (y)
X = data['email']
y = data['label']

# Perform text vectorization
vectorizer = TfidfVectorizer(max_features=1000)  
X_vectorized = vectorizer.fit_transform(X)

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_vectorized, y, test_size=0.2, random_state=42)

# Initialize the XGBoost classifier with suggested hyperparameters
clf = XGBClassifier(
    learning_rate=0.05, 
    n_estimators=500,   
    max_depth=3,        
    subsample=0.9,      
)
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
