# TimeKora.com Web scapping Project
In this file i make a simple web scrapping code for the bootcamp.

```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
# d,m,y =  [int(x) for x in input("Enter three values for the date: (format:DD MM YYYY)").split()] 

d, m, y = "12" ,"08", "2023" # i make a defualt ipnut

page = requests.get(f"https://timekora.com/matches/{y}-{m}-{d}/")

def main(page):
    src = page.content
    soup = BeautifulSoup(src,"lxml")

    matches_details = []
    
    championships = soup.find_all("div",{'class': "matches_container"})
    def get_match_info(champion):
        champion_title = champion.find("div",{'class': "match-title"}).text.strip()
        all_matches = champion.find_all("div",{'class': "match_item single-match-box"})
        matches_count = len(all_matches)
        
        for i in range(matches_count):
            awayTeam = all_matches[i].find("td",{'class': "awayTeam"}).text.strip()
            homeTeam = all_matches[i].find("td",{'class': "homeTeam"}).text.strip()
            result = all_matches[i].find("span",{'class': "match_time"}).text.strip()
#             match_time = all_matches[i].find("span",{'class': "startTime"}).text.strip()
            status = all_matches[i].find("span",{'class': "eventStatus"}).text.strip()
            
            if status == "لم تبدأ":
                result = "-"

            
#             add matches info to matches_details
            matches_details.append({"نوع البطولة":champion_title, "الفريق الأول":homeTeam, "الفريق الثاني":awayTeam, "النتيجة":result})
            
    for i in range(len(championships)):
        get_match_info(championships[i])
    
    keys = matches_details[0].keys()
    with open('matches_details.csv', 'w', encoding='utf-8') as output_file:
        dictwriter = csv.DictWriter(output_file, keys)
        dictwriter.writeheader()
        dictwriter.writerows(matches_details)
        print("done")
    
main(page)
```
